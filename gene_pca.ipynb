```python
x_min, x_max = final_dy.iloc[:, 0].min() - 10000, final_dy.iloc[:, 0].max() + 10000
y_min, y_max = final_dy.iloc[:, 1].min() - 10000, final_dy.iloc[:, 1].max() + 10000
```

```python
#plt.figure(figsize= (15,10))

plt.scatter(final_dy.iloc[:, 0], final_dy.iloc[:, 1], c=final_dy['Target'], cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

plt.show()
```

```python
import matplotlib.cm as cm
x=final_dy['PC1']
y=final_dy['PC2']
classes = final_dy['Target']
cmap = plt.cm.Set1

plt.scatter(x, y, c=classes, label=classes,cmap=cmap)
plt.colorbar()
unique_classes = list(set(classes))
rowcolors=cmap(np.array(unique_classes))
plt.table(cellText=[[x] for x in unique_classes], loc='right',
          colWidths=[0.2],rowColours=rowcolors,
          rowLabels=['label%d'%x for x in unique_classes],
        colLabels=['classes'])
```

```python
plt.scatter(final_dy.iloc[:, 0], final_dy.iloc[:, 1],
            c=final_dy.Target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('spectral',len(set(final_dy.Target))))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

```python
cmap = plt.get_cmap('spectral', 20)
cmap.set_under('gray')

fig, ax = plt.subplots()
cax = ax.scatter(final_dy.iloc[:, 0], final_dy.iloc[:, 1],
            c=final_dy.Target ,s=100, cmap=cmap, vmin=final_dy.Target.min(), vmax=final_dy.Target.max())
fig.colorbar(cax, extend='min')

plt.show()
```
